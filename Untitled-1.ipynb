```python
import requests
import json
```

```python
response = requests.post('http://127.0.0.1:8000/')
response.raise_for_status

json.loads(response.content.decode('utf-8'))
```

```python
file_name = 'sample.jpg'
file = {'img': open(file_name, 'rb')}
headers = {'type': 'multipart/image'}

URL = 'http://127.0.0.1:8000'
filter = "sepia"

response = requests.post("{}/{}".format(URL, filter), files=file, headers=headers)
response.raise_for_status
```

```python
from PIL import Image
import io
import matplotlib.pyplot as plt
```

```python
# plot the image
img = Image.open(io.BytesIO(response.content))
plt.imshow(img)
```

```python
img.save( "{}-{}.{}".format(file_name.split('.')[0], filter,"jpg") , "JPEG" )
```

```python
import aiohttp
import asyncio
import time
```

```python
URL = 'http://127.0.0.1:8000'
start = time.time()
results = []
for i in range(1000):
    results.append(requests.post(URL).content)
end = time.time()
print("Time taken: {}".format(end - start))
```

```python
URLS = [URL for i in range(1000)]

async def test(URL):
    async with aiohttp.ClientSession() as session:
        async with session.post(URL) as response:
            return await response.read()

start = time.time()
loop = asyncio.get_event_loop()
result = await asyncio. gather (*[ loop.create_task(test( URL )) for URL in URLS ], return_exceptions = True )
end = time.time()
print("Time taken: {}".format(end - start))
```
